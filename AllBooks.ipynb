```python
import re
import numpy as np
import pandas as pd
import string

#For Edgar Allen Poe
eap = open("Books/EAP.txt", "r")
eap_text = [ line for line in eap ]
eap_text = eap_text[1:]
eap_text = [string.replace(x,'\n',' ') for x in eap_text]
eap_text = '' . join(eap_text)
eap_sentences = re.split(r'[.!?]', eap_text)
eap_sentences
```

```python
#For HP Lovecraft
hpl = open("Books/hpl.txt", "r")
hpl_text = [ line for line in hpl ]
hpl_text = hpl_text[1:]
hpl_text = [string.replace(x,'\n',' ') for x in hpl_text]
hpl_text = '' . join(hpl_text)
hpl_sentences = re.split(r'[.!?]', hpl_text)
hpl_sentences
```

```python
#For MAry Shelley
ms = open("Books/Mary Shelley/MS.txt", "r")
ms_text = [ line for line in ms ]
ms_text = ms_text[1:]
ms_text = [string.replace(x,'\n',' ') for x in ms_text]
ms_text = '' . join(ms_text)
ms_sentences = re.split(r'[.!?]', ms_text)
ms_sentences
```

```python
#Appending sentences
all_sentences = eap_sentences + hpl_sentences + ms_sentences
```

```python
#Getting rid of punctutations
r=re.compile("[a-zA-Z0-9 ]")
temp=[]
for i in all_sentences:
    i2=filter(r.match,i)
    temp.append(i2.lower())
temp
```

```python
#Creating Bag of words.
import collections
bagofwords = [collections.Counter(re.findall(r'\w+',txt)) for txt in temp]
for i in range(len(bagofwords)):
    bagofwords[i] = dict(bagofwords[i])
bagofwords
```

```python
df = pd.DataFrame(bagofwords)
df
```
