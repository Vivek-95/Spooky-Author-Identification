```python
import pandas as pd
import re
import collections
import numpy as np
train=pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
import re
temp = []
temp_test=[]
r=re.compile("[a-zA-Z0-9 ]")
words={}
for i in train.iloc[:,1]:
    i2=filter(r.match,i)
    temp.append(i2.lower())

for i in test.iloc[:,1]:
    i2=filter(r.match,i)
    temp_test.append(i2.lower())
    
bagofwords = [collections.Counter(re.findall(r'\w+',txt)) for txt in temp]
bagofwords_test=[collections.Counter(re.findall(r'\w+',txt)) for txt in temp_test]
print(bagofwords[1])
```

```python
for i in range(len(bagofwords)):
    bagofwords[i] = dict(bagofwords[i])
for i in range(len(bagofwords_test)):
    bagofwords_test[i] = dict(bagofwords_test[i])
```

```python
new_df = pd.DataFrame(bagofwords)
len(new_df.columns) 
```

```python
new_df2 = pd.DataFrame(bagofwords_test)
len(new_df2.columns)
```

```python
new_df = new_df.fillna(0)
new_df2 = new_df2.fillna(0)
```

```python
import math
temp_df = new_df
temp_df = temp_df.divide(temp_df,axis=1).fillna(0)
sum_of_freq = temp_df.sum(axis=0)
idf = [math.log(len(new_df.columns)/ x) for x in sum_of_freq ]
idf
```

```python
new_df = new_df.mul(np.array(idf),axis=1)
```

```python
temp_test_df = new_df2
temp_test_df = temp_test_df.divide(temp_test_df,axis=1).fillna(0)
sum_of_freq_test = temp_test_df.sum(axis=0)
idf_test = [math.log(len(new_df2.columns)/ x) for x in sum_of_freq_test ]
idf_test
```

```python
new_df2 = new_df2.mul(np.array(idf_test),axis=1)
```

```python
intersection = np.unique(new_df.columns & new_df2.columns)
joineddf = new_df2.loc[:,intersection]
other = [ x for x in new_df.columns if x not in intersection]
```

```python
new_joined = pd.concat([joineddf,pd.DataFrame(columns=other)]).fillna(0)   
```

```python
from sklearn.naive_bayes import MultinomialNB
gnb = MultinomialNB()
pred = gnb.fit(new_df,train.iloc[:,2]).predict_proba(new_joined)
```

```python
final = pd.concat([test.iloc[:,0],pd.DataFrame(pred,columns=['EAP','HPL','MWS'])],axis=1)
```
